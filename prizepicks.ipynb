```python
import requests
import pandas as pd
```

```python

def framer(url,):
    """make sure to implent Rate limit checker and rotate through proxies"""
    headers = {
  "Host": "api.prizepicks.com",
  "User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.9; rv:50.0) Gecko/20100101 Firefox/50.0",
  "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8",
  "Accept-Language": "en-US,en;q=0.5",
  "Accept-Encoding": "gzip, deflate, br",
  "Referer": "https://app.prizepicks.com/",
  "Connection": "keep-alive",
  "Upgrade-Insecure-Requests": "1",
  "Cache-Control": "max-age=0"
}
    response = requests.get(url,headers=headers).json()
    data =pd.json_normalize(response['data'],max_level = 3)
    included = pd.json_normalize(response['included'], max_level=3)
    inc_cop = included[included['type'] == 'new_player'].copy().dropna(axis=1)
    data = pd.merge(data, inc_cop, how='left', left_on=['relationships.new_player.data.id','relationships.new_player.data.type']
                        , right_on=['id', 'type']
                        , suffixes=('', '_new_player')).set_index('id')

    return data
    
```

```python
url =  'https://api.prizepicks.com/projections?&per_page=250&single_stat=true'
data = framer(url)


```

```python
common_indices = data.index.intersection(current_data.index)
changed_column = current_data.loc[common_indices, 'attributes.line_score'] != data.loc[common_indices, 'attributes.line_score']
changed_indices = changed_column[changed_column].index
changed_rows = current_data.loc[changed_indices]
print(changed_rows)
```

```python
historical_changes = {}

for index,row in changed_rows.iterrows():
    old_row = data.loc[index]
    if index not in historical_changes:
        historical_changes[index] = []
        changed_lines = f"{row['attributes.name']}:{old_row['attributes.line_score']} to {row['attributes.line_score']}"
        historical_changes[index].append(changed_lines)
    else:
        historical_changes[index].append( f" to {row['attributes.line_score']}")

    print(historical_changes[index])
    
```
